```python
#15.Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
#tasks using Matplotlib:
#a) Plot a histogram of the data with 30 bins.

#b) Overlay a line plot representing the normal distribution's probability density function (PDF).

#c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

#d) Set the title of the plot as 'Histogram with PDF Overlay'.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

data = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(data, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
pdf = norm.pdf(x, loc=0, scale=1)

plt.plot(x, pdf, 'k', linewidth=2)

plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')

plt.show()
```

```python
#16.Set the title of the plot as 'Histogram with PDF Overlay'.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

data = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(data, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(data), np.std(data))
plt.plot(x, p, 'k', linewidth=2)

plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

plt.title('Histogram with PDF Overlay')

plt.show()
```

```python
#17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)

def get_quadrant(x, y):
    if x >= 0 and y >= 0:
        return 'Quadrant 1'
    elif x < 0 and y >= 0:
        return 'Quadrant 2'
    elif x < 0 and y < 0:
        return 'Quadrant 3'
    else:
        return 'Quadrant 4'

df = pd.DataFrame({
    'X': x,
    'Y': y,
    'Quadrant': [get_quadrant(xi, yi) for xi, yi in zip(x, y)]
})

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='X', y='Y', hue='Quadrant', palette='viridis', s=100)


plt.legend(title='Quadrant')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

plt.show()
```

```python
#18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

p = figure(title='Sine Wave Function', x_axis_label='X', y_axis_label='Y', width=800, height=400)

p.line(x, y, line_width=2, color='blue', legend_label='Sine Wave')

p.grid.grid_line_color = 'gray'
p.grid.grid_line_alpha = 0.5

show(p)

```

```python
#19.Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
#values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import linear_cmap
import pandas as pd
import numpy as np

output_notebook()

categories = [f'Category {i}' for i in range(10)]
values = np.random.randint(1, 100, size=len(categories))

data = pd.DataFrame({'Category': categories, 'Value': values})

source = ColumnDataSource(data=data)

color_mapper = linear_cmap(field_name='Value', palette='Viridis256', low=min(values), high=max(values))

p = figure(x_range=categories, title='Random Categorical Bar Chart',
           x_axis_label='Category', y_axis_label='Value', width=800, height=400)

p.vbar(x='Category', top='Value', width=0.9, source=source,
       line_color='white', fill_color=color_mapper)

hover = HoverTool()
hover.tooltips = [("Category", "@Category"), ("Value", "@Value")]
p.add_tools(hover)

p.grid.grid_line_color = 'gray'
p.grid.grid_line_alpha = 0.5

show(p)

```

```python
#20.Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'
```

```python
import plotly.graph_objects as go
import numpy as np

np.random.seed(0)
x = np.arange(100)
y = np.random.randn(100)

fig = go.Figure()

fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))

fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X-axis',
    yaxis_title='Y-axis'
)

fig.show()
```

```python
#21.Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set the title as 'Interactive Pie Chart'.
```

```python
import plotly.graph_objects as go
import numpy as np


np.random.seed(0)
categories = [f'Category {i}' for i in range(10)]
values = np.random.randint(1, 100, size=len(categories))

fig = go.Figure(data=[go.Pie(
    labels=categories,
    values=values,
    hole=0.3,
    textinfo='label+percent',
    title='Interactive Pie Chart'
)])

fig.update_layout(
    title='Interactive Pie Chart'
)


fig.show()
```
