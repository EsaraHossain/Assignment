```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.sampledata.iris import flowers
from bokeh.plotting import figure, output_file, show
```

```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
```

```python

x = np.linspace(0, 10, 500)
y = np.sin(x)
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, legend_label="sin(x)", line_width=2, color="blue")
output_file("sine_wave.html")

show(p)
```

```python
#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns
```

```python

np.random.seed(42)
n = 100
x = np.random.rand(n) * 10
y = np.random.rand(n) * 10
sizes = np.random.rand(n) * 20
colors = np.random.rand(n, 3)

colors_hex = ['#%02x%02x%02x' % (int(r * 255), int(g * 255), int(b * 255)) for r, g, b in colors]

data = {
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': colors_hex
}
df = pd.DataFrame(data)

p = figure(title="Scatter Plot with Random Data", x_axis_label='x', y_axis_label='y')
p.scatter(x='x', y='y', size='sizes', color='colors', source=df, legend_label="Random Points", fill_alpha=0.6)
output_file("scatter_plot.html")
show(p)
```

```python
#3.Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
```

```python
import pandas as pd
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook

data = {
    'fruits': ['Apples', 'Oranges', 'Bananas', 'Pears'],
    'counts':  [20, 25, 30, 35]

}

df = pd.DataFrame(data)

p = figure(x_range=df['fruits'], title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count',
           toolbar_location=None, tools="", height=350, width=600)


p.vbar(x=df['fruits'], top=df['counts'], width=0.9, color='green', legend_label="Counts")

p.xgrid.grid_line_color = None
p.y_range.start = 0
p.axis.minor_tick_line_color = None
p.outline_line_color = None
p.xaxis.major_label_orientation = 1.2

output_file("fruit_counts.html")

output_notebook()
show(p)
```

```python
#4.Create a Bokeh histogram to visualize the distribution of the given data
```

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook


data_hist = np.random.randn(1000)

hist, edges = np.histogram(data_hist, bins=30)

df = pd.DataFrame({
    'left_edges': edges[:-1],
    'right_edges': edges[1:],
    'counts': hist
})
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency',
           toolbar_location=None, tools="", height=350, width=600)

p.quad(top='counts', bottom=0, left='left_edges', right='right_edges', source=df, fill_color="navy", line_color="white", alpha=0.7)

p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None

output_file("histogram.html")

output_notebook()
show(p)
```

```python
#5.Create a Bokeh heatmap using the provided dataset.
```

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook
from bokeh.transform import LinearColorMapper
from bokeh.palettes import Viridis256

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

df_heatmap = pd.DataFrame({
    'x': xx.flatten(),
    'y': yy.flatten(),
    'value': data_heatmap.flatten()
})

color_mapper = LinearColorMapper(palette=Viridis256, low=df_heatmap['value'].min(), high=df_heatmap['value'].max())

p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1), toolbar_location=None, tools="",
           width=600, height=600, tooltips=[("x", "@x"), ("y", "@y"), ("value", "@value")])

p.rect(x='x', y='y', width=0.1, height=0.1, source=df_heatmap,
       fill_color={'field': 'value', 'transform': color_mapper},
       line_color=None, hover_fill_color='white', hover_line_color='black')

p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
p.xaxis.major_label_orientation = 0
p.yaxis.major_label_orientation = 0
output_file("heatmap.html")

output_notebook()
show(p)

```
